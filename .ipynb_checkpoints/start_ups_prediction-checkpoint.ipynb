```python
#     
date_columns = ['first_funding_at', 'last_funding_at']
#           
def transform_dates(df, date_columns):
    #      
    for col in date_columns:
        #     datetime
        df[col] = pd.to_datetime(df[col], errors='coerce')
        
        #       
        df[f'{col}_year'] = df[col].dt.year
        df[f'{col}_month'] = df[col].dt.month
    
    #     
    df = df.drop(columns=date_columns)
    
    return df
```

```python
#     
def calculate_lifetime(df):
    #       datetime
    df['founded_at'] = pd.to_datetime(df['founded_at'], errors='coerce')
    df['closed_at'] = pd.to_datetime(df['closed_at'], errors='coerce')

    #    ,  'closed_at'  (2018-01-01)
    default_date = pd.to_datetime('2018-01-01')
    
    #  'closed_at' ,  default_date
    df['closed_at'] = df['closed_at'].fillna(default_date)

    #      
    df['lifetime'] = (df['closed_at'] - df['founded_at']).dt.days

    #   founded_at
    df = df.drop(columns=['founded_at', 'closed_at'])
    
    return df
```

```python
#    
def clean_text(text):
    if isinstance(text, str):  # ,   
        #   ,         /
        return ' '.join(text.split()).strip().lower()
    return text  #   ,     (, NaN)
```

```python
def cat_consolidation(original_df, category_mapping):
    #  0:    
    df = original_df.copy()    
    
    #    ,    'Initial',    'Final'
    mapping = {row['Initial']: row['Final'] for _, row in category_mapping.iterrows()}
    
    #  2:     |,  
    df['categories_split'] = df['category_list'].fillna('unknown').str.lower().str.split('|')  #    
    
    #  3:         
    df['mapped_categories'] = df['categories_split'].apply(
        lambda categories: list(set([mapping.get(cat.strip(), cat) for cat in categories])) if isinstance(categories, list) else []
    )
    
    #  4:     
    mlb = MultiLabelBinarizer()
    binary_categories = pd.DataFrame(mlb.fit_transform(df['mapped_categories']), columns=mlb.classes_, index=df.index)
    
    #  5:       
    columns_to_add = [col for col in binary_categories.columns if col not in df.columns]
    
    #      
    df = df.join(binary_categories[columns_to_add])
    
    #  6:   ,      ,  
    all_final_categories = set(mapping.values())
    missing_categories = all_final_categories - set(binary_categories.columns)
    
    #     
    for category in missing_categories:
        df[category] = 0
    
    #  7:      
    df = df.drop(columns=['category_list', 'categories_split', 'mapped_categories'])
    
    #   DataFrame
    return df
```

```python
#     
X = startups_train_cons_cat.drop(columns=['status'])
y = startups_train_cons_cat['status']
```

```python
#     
numeric_features = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = X.select_dtypes(include=['category']).columns.tolist()
```

```python
#      
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,  #    (, 20%)
    random_state=RANDOM_STATE,
    stratify=y  #    
)
```

```python
#   
le = LabelEncoder()
y_train_encoded = le.fit_transform(y_train)
y_test_encoded = le.transform(y_test)
```

```python
#       
imputation_pipe = Pipeline([
    ('imputer_nan', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
    ('imputer_space', SimpleImputer(missing_values=' ', strategy='most_frequent'))
])
```

```python
#   One-Hot Encoding
ohe_pipe = Pipeline([
    ('imputation', imputation_pipe),
    ('ohe', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False))
])

#    
numeric_pipe = Pipeline([
    ('imputer_nan', SimpleImputer(missing_values=np.nan, strategy='mean')),
    ('scaler', StandardScaler())
])

#    
data_preprocessor_ohe = ColumnTransformer([
    ('num', numeric_pipe, numeric_features),
    ('ohe', ohe_pipe, categorical_features)
], remainder='passthrough')

#      OHE
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor_ohe),
    ('feature_selection', SelectKBest(score_func=f_classif)), #  SeleckKBest
    ('model', LogisticRegression())  #    
])
```

```python
#    
param_distributions_lr = [
    {
#         'preprocessor': [data_preprocessor_ohe],
        'model': [LogisticRegression(solver='liblinear', random_state=RANDOM_STATE)],
        'model__C': [1, 2, 3, 4, 5, 10],
        'feature_selection__k': list(range(5, len(X_train.columns) + 1))
    }
]
    

param_distributions_knnc = [   
    {
        'model': [KNeighborsClassifier()],
        'model__n_neighbors': [2, 4],
        'feature_selection__k': list(range(5, len(X_train.columns) + 1))
    }
]

param_distributions_treec = [ 
    {
        'model': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'model__max_depth': [10, 20, 30],
        'feature_selection__k': list(range(5, len(X_train.columns) + 1))
    }
]

param_distributions_svc = [
    {
        'model': [SVC(probability=True, random_state=RANDOM_STATE)],
        'model__kernel': ['linear', 'poly', 'rbf'],
        'model__C': [0.1, 1, 10],
        'feature_selection__k': list(range(5, len(X_train.columns) + 1))
    }
]
```

```python
#     precision, recall  F1-score
def calculate_custom_f1(y_true, y_pred, pos_label=1):
    #   
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()

    #  Precision  Recall
    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0

    #  F1-Score
    if precision + recall > 0:
        f1 = 2 * (precision * recall) / (precision + recall)
    else:
        f1 = 0
    
    return precision, recall, f1
```

```python
#    
def randomized_search_f1(param_distributions, pos_label='closed'):
    random_search = RandomizedSearchCV(
        pipe_final, 
        param_distributions, 
        n_iter=10,  #    
        scoring='f1_samples',  #  F1 
        cv=KFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE),
        random_state=RANDOM_STATE,
        n_jobs=-1  #  
    )
    #     
    random_search.fit(X_train, y_train)
    #     
    best_pipe = random_search.best_estimator_
    
    #     
    y_pred_numeric = best_pipe.predict(X_test)

    #       ('closed', 'operating')
    y_pred = le.inverse_transform(y_pred_numeric)

    #   
    precision, recall, f1 = calculate_custom_f1(y_test_encoded, y_pred, pos_label=le.transform(['closed'])[0])

    #   
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-Score: {f1:.4f}")

    # 
    results_df = pd.DataFrame({
        'Model': [best_pipe.named_steps['model'].__class__.__name__],
        'Precision': [precision],
        'Recall': [recall],
        'F1-Score': [f1]
    })

    best_params_dict = {
        best_pipe.named_steps['model'].__class__.__name__: random_search.best_params_
    }

    return results_df, best_params_dict
```
